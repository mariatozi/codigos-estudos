```python
def substituir_valor(texto):
    if texto in ['LIGAÇÃO MUDA - SEM RETORNO DE VOZ']:
        return 'Ligação sem retorno de voz'
    elif texto in ['Durante o atendimento: Ligação caiu/Sem retorno de voz']:
        return 'Ligação caiu/sem retorno durante atendimento'
    elif texto in ['Falha(*): queda na ligação']:
        return 'Queda na ligação'
    elif texto in ['Falha(*):ligação muda']:
        return 'Ligação muda'
    elif texto in ['Erro no Venus']:
        return 'Fluxo de falha sistêmica'
    else:
        return 'verificar'
```

```python
df['Desc-dados'] = df['Descrição'].apply(substituir_valor)
```
